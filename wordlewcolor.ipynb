```python
import tkinter as tk
from tkinter import *

def insert_text(letter):
    current_text = textbox.get("1.0", "end-1c")
    new_text = current_text + letter
    textbox.delete("1.0", "end")
    textbox.insert("1.0", new_text)

def Backspace():
    textbox.delete("end-2c", "end")
def ClearText():
    textbox.delete("1.0", "end")

turn=1
guess=""
def Actual_Game():
    global guess
    global turn
    actualWord = "QUALM"
    guess = textbox.get("1.0", "end-1c").upper().strip()
    gylist = ['#000000','#000000','#000000','#000000','#000000']

    if len(guess)!=5:
        LenMessage_label.config(text="Only 5 Letter Words!")
    else:
    
        # Find all green letters
        if turn <=5:
            for gcNum, guessChar  in enumerate(guess):
                for acNum, acChar in enumerate(actualWord):
                    if (guessChar == acChar and gcNum == acNum):
                        gylist[gcNum] = '#6ca965'
            
            
            for acNum, acChar in enumerate(actualWord):
                if (gylist[acNum] != '#6ca965'):
                    for gcNum, guessChar  in enumerate(guess):
                        if (guessChar == acChar and gylist[gcNum] != '#6ca965'):
                            gylist[gcNum] = '#c8b653'
                
                            # To prevent guess letters from being marked yellow for the same character, break inner-loop
                            break


                





    
    GuessBtns = tk.Frame(root)
    GuessBtns.columnconfigure(3, weight=1)
    GuessBtns.columnconfigure(4, weight=1)
    GuessBtns.columnconfigure(5, weight=1)
    
    for k in range(0,5):
        GuessLetter = tk.Label(GuessBtns, text= guess[k], font=('Arial',25),bg=gylist[k])
        GuessLetter.grid(row=turn, column=k, sticky=tk.W+tk.E, padx=5, pady=5)
        GuessBtns.pack()
    def check(list):
        return all(i == list[0] for i in list)
    if check(gylist) is True and gylist[0]=='#6ca965':
        lose_win_label.config(text="You win!")
        btnSubmit.config(state=DISABLED)
    if check(gylist) is False and turn ==5:
        lose_win_label.config(text="You Lose! The word was:".format(actualWord))
        btnSubmit.config(state=DISABLED)
        
          
    turn +=1
    ClearText()


    
root = tk.Tk()
root.title("Bootleg Wordle")
root.geometry("600x800")

label = tk.Label(root, text='Please Enter a Guess!', font=('Arial',18))
label.pack(pady=20)

textbox = tk.Text(root, height=3, font=('Arial',18))
textbox.pack(padx=20)

buttonframe = tk.Frame(root)
buttonframe.columnconfigure(0, weight=1)
buttonframe.columnconfigure(1, weight=1)
buttonframe.columnconfigure(2, weight=1)

Alphabet = "QWERTYUIOPASDFGHJKLZXCVBNM"


for index, letter in enumerate(Alphabet):
    btnltr = tk.Button(buttonframe, text=letter, font=('Arial', 18), command=lambda l=letter: insert_text(l))
    if index <=9:
        btnltr.grid(row=0, column=index, sticky=tk.W + tk.E)

    elif index >=10 and index <=18:
        btnltr.grid(row=1, column=index-10, sticky=tk.W + tk.E)

    else:
        btnltr.grid(row=2, column=index-19, sticky=tk.W + tk.E)

        
buttonframe.pack()





btnfrm2 = tk.Frame(root)
btnfrm2.columnconfigure(0, weight=1)
btnfrm2.columnconfigure(1, weight=1)
btnfrm2.columnconfigure(2, weight=1)

btnDelete = tk.Button(btnfrm2, text='Delete', font=('Arial',18),command=Backspace)
btnDelete.grid()

btnSubmit = tk.Button(btnfrm2, text ='Guess',font=('Arial',18),command=Actual_Game)
btnSubmit.grid()
btnfrm2.pack()


LenMessage_label = tk.Label(root, text="", font=('Arial', 18))
LenMessage_label.pack()

guess1_label = tk.Label(root, text="", font=('Arial', 1))
guess1_label.pack()

guess2_label = tk.Label(root, text="", font=('Arial', 1))
guess2_label.pack()

guess3_label = tk.Label(root, text="", font=('Arial', 1))
guess3_label.pack()

guess4_label = tk.Label(root, text="", font=('Arial', 1))
guess4_label.pack()

guess5_label = tk.Label(root, text="", font=('Arial', 1))
guess5_label.pack()

lose_win_label = tk.Label(root, text= "", font=('Arial',24))
lose_win_label.pack()


# Start the main event loop
root.mainloop()
```
